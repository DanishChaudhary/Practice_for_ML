```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sample_data/anime.csv')
```

```python
def episode(text):
  data = ""
  check = False
  for i in text:
    if i == ")":
      check = False
    if i == '(':
      check  = True
    if check:
      data += i
  return data
```

```python
df['Episodes'] = df["Title"].apply(episode)
```

```python
df.head()
```

```python
df['Episodes'] = df['Episodes'].str.replace(r'\)|\(|eps','', regex=True).astype(int)
```

```python
df['Episodes'].head().dtype
```

```python
df.loc[6]["Title"]
```

```python
def time_stamp(text):
  check = False
  data = ''
  for i in range(len(text)):
    if text[i] == ')':
      for j in range(i+1, i+20):
        data += text[j]
      return data
```

```python
df['Time_Stamp'] = df['Title'].apply(time_stamp)
```

```python
df['TotalMonths'] = (
    pd.to_datetime(df['Time_Stamp'].str.split(' - ').str[1], format='%b %Y').dt.year * 12 +
    pd.to_datetime(df['Time_Stamp'].str.split(' - ').str[1], format='%b %Y').dt.month
) - (
    pd.to_datetime(df['Time_Stamp'].str.split(' - ').str[0], format='%b %Y').dt.year * 12 +
    pd.to_datetime(df['Time_Stamp'].str.split(' - ').str[0], format='%b %Y').dt.month
)
```
