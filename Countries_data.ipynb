```python
#which country has the highest population
df[df['population'].max() == df['population']]['country']
```

```python
#which country has the least population
df[df['population'].min() == df['population']]['country']
```

```python
#how many country have Republic in their full name
def find(text):
    if 'Republic' in text:
        return 1
    else:
        return 0
```

```python
df['country_long'].apply(find).sum()
```

```python
df['country_long'].str.contains('Republic', case=False).sum()
```

```python
#which country in african region has highest population
df[df['continent'] == 'Africa'].sort_values('population', ascending=False).head(1)['country']
```
